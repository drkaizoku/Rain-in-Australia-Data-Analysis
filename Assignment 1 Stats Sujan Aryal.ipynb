```python
#Loading Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

WeatherData = pd.read_csv('weatherAUS.csv',delimiter=',',)
```

```python
# Finding the Mean maximum temperature
mean_max_temp = WeatherData['MaxTemp'].mean()
print(f"Mean Maximum Temperature: {mean_max_temp:.2f}Â°C")
```

```python
# Finding the Mean minimum temperature
mean_min_temp = WeatherData['MinTemp'].mean()
print(f"Mean Minimum Temperature: {mean_min_temp:.2f}Â°C")
```

```python
# Plotting the Histogram for Maximum Temperature
plt.figure(figsize=(10, 6))
sns.distplot(WeatherData['MaxTemp'], kde=True, bins=30, color='blue')
plt.title('Histogram of Maximum Temperature')
plt.xlabel('Maximum Temperature (Â°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the Histogram for Minimum Temperature
plt.figure(figsize=(10, 6))
sns.distplot(WeatherData['MinTemp'], kde=True, bins=30, color='green')
plt.title('Histogram of Minimum Temperature')
plt.xlabel('Minimum Temperature (Â°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the Boxplot for Maximum and Minimum Temperatures
plt.figure(figsize=(10, 6))
sns.boxplot(data=WeatherData[['MaxTemp', 'MinTemp']], palette='pastel')
plt.title('Boxplot of Maximum and Minimum Temperatures')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

```python
# Comparing which year has the most rain from the weather Datset

# Converting 'Date' column to datetime format
WeatherData['Date'] = pd.to_datetime(WeatherData['Date'])

# Extracting year from 'Date' column
WeatherData['Year'] = WeatherData['Date'].dt.year
```

```python
# Grouping data by year and sum the rainfall
yearly_rain = WeatherData.groupby('Year')['Rainfall'].sum().reset_index()
```

```python
# Now Finding the year with the maximum rainfall
max_rain_year = yearly_rain.loc[yearly_rain['Rainfall'].idxmax()]
print(f"The year with the most rain is {int(max_rain_year['Year'])} with {max_rain_year['Rainfall']:.2f} mm of rainfall.")
```

```python
# Data Visualization of the year which has the most rain fall in 10 years

plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Rainfall', data=yearly_rain, palette='viridis')
plt.title('Total Rainfall by Year')
plt.xlabel('Year')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Checking data types of the columns
print(WeatherData.dtypes)
```

```python
# Convert object types to float
WeatherData['MaxTemp'] = WeatherData['MaxTemp'].astype(float)
WeatherData['MinTemp'] = WeatherData['MinTemp'].astype(float)
WeatherData['Rainfall'] = WeatherData['Rainfall'].astype(float)
```

```python
# Drop rows with missing values in the selected columns
WeatherData = WeatherData.dropna(subset=['MaxTemp', 'MinTemp', 'Rainfall'])
```

```python
# Scatter plot of MaxTemp vs Rainfall using matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(WeatherData['MaxTemp'], WeatherData['Rainfall'])
plt.title('Max Temperature vs Rainfall')
plt.xlabel('Max Temperature (Â°C)')
plt.ylabel('Rainfall (mm)')
plt.show()

# Scatter plot of MinTemp vs Rainfall using matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(WeatherData['MinTemp'], WeatherData['Rainfall'])
plt.title('Min Temperature vs Rainfall')
plt.xlabel('Min Temperature (Â°C)')
plt.ylabel('Rainfall (mm)')
plt.show()

```

```python
# Check for missing values
print(WeatherData[['MaxTemp', 'MinTemp', 'Rainfall']].isnull().sum())
# Drop rows with missing values
WeatherData = WeatherData.dropna(subset=['MaxTemp', 'MinTemp', 'Rainfall'])
```

```python
print(WeatherData[['MaxTemp', 'MinTemp', 'Rainfall']].head())
```

```python
# Calculate mean rainfall for each location
mean_rainfall = WeatherData.groupby('Location')['Rainfall'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(15, 10))
mean_rainfall.plot(kind='bar', color='green')
plt.title('Average Rainfall by Location')
plt.xlabel('Location')
plt.ylabel('Average Rainfall (mm)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group data by location to sum up daily rainfall
LocationRainfall = WeatherData.groupby('Location')['Rainfall'].sum().sort_values()

# Create the bar chart
plt.figure(figsize=(15, 10))
LocationRainfall.plot(kind='barh', color='blue')
plt.title('Total Rainfall by Location (2007 to 2017)')
plt.xlabel('Total Rainfall (mm)')
plt.ylabel('Location')
plt.grid(axis='x')
plt.show()
```

```python
# Group data by year to sum up total rainfall
YearlyRainfall = WeatherData.groupby('Year')['Rainfall'].sum()

# Create a line chart
plt.figure(figsize=(12, 6))
YearlyRainfall.plot(kind='line', marker='o', color='red')
plt.title('Yearly Rainfall Trend (2007 to 2017)')
plt.xlabel('Year')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Display column names in the dataset
print(WeatherData.columns)
```

```python
# Filter data for heavy rainfall areas (rainfall > 50mm)
HeavyRainfallData = WeatherData[WeatherData['Rainfall'] > 50]

# Group filtered data by location to sum up total rainfall
HeavyRainfallLocation = HeavyRainfallData.groupby('Location')['Rainfall'].sum().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(15, 6))
HeavyRainfallLocation.plot(kind='bar', color='purple')
plt.title('Total Heavy Rainfall (>50mm) by Location (2007 to 2017)')
plt.xlabel('Location')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Group data by year to calculate mean values
YearlyData = WeatherData.groupby('Year').agg({
    'Rainfall': 'mean',
    'MinTemp': 'mean',
    'MaxTemp': 'mean',
    'Evaporation': 'mean'
})

# Create the line graph
plt.figure(figsize=(12, 6))
plt.plot(YearlyData.index, YearlyData['Rainfall'], marker='o', label='Rainfall', color='blue')
plt.plot(YearlyData.index, YearlyData['MinTemp'], marker='o', label='MinTemp', color='green')
plt.plot(YearlyData.index, YearlyData['MaxTemp'], marker='o', label='MaxTemp', color='red')
plt.plot(YearlyData.index, YearlyData['Evaporation'], marker='o', label='Evaporation', color='orange')

# Setting plot title and labels for visualization
plt.title('Weather Trends Over the Years (2007 to 2017)')
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.xticks(YearlyData.index)
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
# Group data by season to sum up total rainfall
SeasonalRainfall = WeatherData.groupby('Season')['Rainfall'].sum()

# Create a bar chart
plt.figure(figsize=(5, 6))
SeasonalRainfall.plot(kind='bar', color='orange')
plt.title('Seasonal Rainfall (2007 to 2017)')
plt.xlabel('Season')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
